```python
from IPython.utils import io
import time
```

```python
for title in ['Talk:Disneyland Railroad', 'Talk:Raksha Bandhan', 'User talk:Jackdude101', 'User talk:Bondegezou', 'Talk:Chick McGee']:
    namespace, history = get_revisions(title)
    processor = Conversation_Constructor(COMMENT_TRACKING_FILE = "comments.json")
    actions_lst = []
    start = time.time()
    for ind, h in enumerate(history):
        DEBUGGING_MODE = True
        actions = processor.process(h, DEBUGGING_MODE = False)
        for action in actions:
            actions_lst.append(action)
    processor.save("processor.json")
    convs, conv_mapping = group_into_conversations(history[0]['page_id'], actions_lst)
    with open("sample_outputs/%s_conversations"%(title), "w") as w:
        for key, conv in convs.items():
            w.write('CONVERSATION %s\n'%(key))
            snapshots = generate_snapshots(conv)
            with io.capture_output() as captured:
                display(snapshots[-1])
            w.write(captured.stdout)
    print(title, time.time() - start)
```
